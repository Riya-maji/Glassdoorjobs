Converting the data type of salary column from 'object' to 'integer'

```python
salary = df["Salary Estimate"].apply(lambda x : x.split('(')[0])
salary
```

```python
salary = salary.apply(lambda x : x.replace("$" , "").replace("K" , ""))
```

```python
salary
```

Storing the range of values intp Minimum and Maximum categories

```python
df[" Min _salary in K"] = salary.apply(lambda x : x.split("-")[0])
```

```python
df[" Min _salary in K"]
```

```python
df[" Max _salary in K"] = salary.apply(lambda x : x.split("-")[1])
df[" Max _salary in K"]
```

```python
df[" Min _salary in K"] = df[" Min _salary in K"].astype(int)
```

```python
df[" Max _salary in K"] = df[" Max _salary in K"].astype(int)
```

```python
df["avg_salary"] = (df[" Min _salary in K"]+ df[" Max _salary in K"])/2
```
